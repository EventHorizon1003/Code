```python
# === EE5907 Assignment 2: CMU PIE + Selfie Dataset Prep & Visualization ===
# Folder layout expected:
#   CMU_PIE_ROOT/
#     1/   -> 1.jpg, 2.jpg, ...
#     2/   -> ...
#     ...
#     (up to 68 subjects; some datasets may contain 63see MAX_SID)
#
# Selfies: a folder with >=10 images (any names). We take first 10 after natural sorting.

import os, re, math, random
from typing import List, Tuple, Dict
from pathlib import Path

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# ---------------- CONFIG (EDIT THESE) ----------------
CMU_PIE_ROOT = Path(r"")      # e.g., ".../CMU_PIE"
SELFIE_DIR   = Path(r"/path/to/selfies")      # e.g., ".../selfies"
SAVE_DIR     = Path("./figures")              # where to save generated figures
SAVE_DIR.mkdir(parents=True, exist_ok=True)

LAST3      = 567                              # <-- replace with your last 3 matric digits
IMAGE_SIZE = (32, 32)
TRAIN_RATIO = 0.7
MAX_SID    = 68                               # if your set is 1..63, change to 63

# ----------------------------------------------------

# ---------------- Utilities ----------------
def set_seed(seed: int):
    random.seed(seed)
    np.random.seed(seed)

def select_subjects(seed: int, k: int = 25) -> List[int]:
    set_seed(seed)
    return random.sample(range(1, MAX_SID + 1), k)

def _parse_num_from_stem(stem: str) -> int:
    # Extract first number in the filename stem
    m = re.search(r"\d+", stem)
    return int(m.group(0)) if m else -1

def to_gray_resized(im_path: Path, size=(32,32)) -> np.ndarray:
    im = Image.open(im_path).convert("L").resize(size, Image.BILINEAR)
    return np.asarray(im, dtype=np.uint8)

def preprocess_images(paths: List[Path], size=(32,32)) -> np.ndarray:
    if not paths:
        raise ValueError("preprocess_images: empty path list.")
    X = np.stack([to_gray_resized(p, size) for p in paths], axis=0)   # [N,H,W]
    assert X.ndim == 3, f"Expect [N,H,W], got {X.shape}"
    return X

def split_indices(n: int, train_ratio: float, seed: int) -> Tuple[np.ndarray, np.ndarray]:
    rng = np.random.default_rng(seed)
    idx = np.arange(n)
    rng.shuffle(idx)
    ntr = int(round(train_ratio * n))
    if n >= 2:
        ntr = max(1, min(n-1, ntr))  # ensure both splits non-empty for n>=2
    return idx[:ntr], idx[ntr:]

# ---------------- CMU PIE loaders (folders: "1","2",... each with .jpg files) ----------------
def find_subject_images_numeric(root: Path, sid: int) -> Tuple[List[Path], List[int]]:
    """Return (paths, file_numeric_ids) sorted by numeric id from filename (e.g., 1.jpg -> 1)."""
    subj_dir = root / f"{sid}"
    if not subj_dir.is_dir():
        return [], []
    paths = list(subj_dir.glob("*.jpg"))  # JPG only as requested
    pairs = [(p, _parse_num_from_stem(p.stem)) for p in paths]
    pairs = [(p, n) for (p, n) in pairs if n >= 0]
    pairs.sort(key=lambda t: t[1])
    if not pairs:
        return [], []
    paths_sorted, nums_sorted = zip(*pairs)
    return list(paths_sorted), list(nums_sorted)

def load_cmu_pie_block(root: Path, subjects: List[int], seed: int,
                       size=(32,32), train_ratio=0.7):
    """
    Returns:
      Xtr_list, ytr_list, Xte_list, yte_list,
      paths_tr_list, ids_tr_list, paths_te_list, ids_te_list
    where ids_* are numeric file IDs (1..170) for titles like S<sid>_<num>.
    """
    Xtr_list, ytr_list, Xte_list, yte_list = [], [], [], []
    paths_tr_list, ids_tr_list, paths_te_list, ids_te_list = [], [], [], []

    for sid in subjects:
        paths, nums = find_subject_images_numeric(root, sid)
        if not paths:
            raise FileNotFoundError(f"No .jpg images for subject {sid} under {root}")

        X = preprocess_images(paths, size=size)  # [N,H,W]
        ti, vi = split_indices(len(X), train_ratio, seed + sid)

        Xtr, Xte = X[ti], X[vi]
        ytr = np.full((len(ti),), -1, dtype=int)  # temp labels; relabel later
        yte = np.full((len(vi),), -1, dtype=int)

        Xtr_list.append(Xtr);  ytr_list.append(ytr)
        Xte_list.append(Xte);  yte_list.append(yte)

        paths_tr_list.append([paths[i] for i in ti])
        paths_te_list.append([paths[i] for i in vi])
        ids_tr_list.append([nums[i] for i in ti])
        ids_te_list.append([nums[i] for i in vi])

    return (Xtr_list, ytr_list, Xte_list, yte_list,
            paths_tr_list, ids_tr_list, paths_te_list, ids_te_list)

# ---------------- Selfie loaders ----------------
def list_images_numeric(dirpath: Path) -> List[Path]:
    exts = ("*.png","*.jpg","*.jpeg","*.bmp","*.tif","*.tiff")
    all_imgs = [p for ext in exts for p in dirpath.glob(ext)]
    # natural-ish numeric-first sort on stem
    def keyfun(p):
        try:
            return (0, int(p.stem))
        except:
            return (1, p.stem)
    return sorted(all_imgs, key=keyfun)

def load_selfies_block(selfie_dir: Path, size=(32,32)) -> Tuple[np.ndarray, List[Path]]:
    all_imgs = list_images_numeric(selfie_dir)
    if len(all_imgs) < 10:
        raise ValueError(f"Need at least 10 selfies in {selfie_dir}, found {len(all_imgs)}.")
    sel10 = all_imgs[:10]
    X = preprocess_images(sel10, size=size)
    return X, sel10

def split_selfies(Xself: np.ndarray) -> Tuple[np.ndarray, np.ndarray]:
    if Xself.shape[0] != 10:
        raise ValueError("Selfie block must contain exactly 10 images after preprocessing.")
    return Xself[:7], Xself[7:]  # 7 train, 3 test

# ---------------- Label map & stacking ----------------
def make_label_map(subjects: List[int]) -> Dict[str, int]:
    # 0..24 for CMU PIE 25 subjects; 25 for selfie
    labmap = {f"subject_{sid:02d}": i for i, sid in enumerate(subjects)}
    labmap["selfie"] = 25
    return labmap

def relabel(y: np.ndarray, new_label: int) -> np.ndarray:
    y[:] = new_label
    return y

def stack_xy(blocks: List[np.ndarray], labels: List[np.ndarray]) -> Tuple[np.ndarray, np.ndarray]:
    if not blocks:
        return np.empty((0, IMAGE_SIZE[1], IMAGE_SIZE[0]), dtype=np.uint8), np.empty((0,), dtype=int)
    X = np.concatenate(blocks, axis=0)
    y = np.concatenate(labels, axis=0)
    return X, y

# ---------------- Plots ----------------
def show_selfies_grid(Xself: np.ndarray, paths: List[Path], cols=5, save_path: Path=None):
    rows = math.ceil(len(Xself) / cols)
    plt.figure(figsize=(cols*2, rows*2))
    for i, (img, p) in enumerate(zip(Xself, paths)):
        plt.subplot(rows, cols, i+1)
        plt.imshow(img, cmap="gray")
        plt.title(p.name, fontsize=8)
        plt.axis("off")
    plt.suptitle("Your 10 selfies (grayscale, 3232)")
    plt.tight_layout()
    if save_path: plt.savefig(save_path, dpi=200)
    plt.show()

def show_cmu_pie_examples_labeled(
    X_blocks: List[np.ndarray],
    subjects: List[int],
    ids_blocks: List[List[int]],
    subject_indices: List[int]=None,
    imgs_per_subject: int=5,
    save_path: Path=None
):
    """Plots train images for chosen subject indices, title as S<sid>_<num>."""
    if subject_indices is None:
        subject_indices = list(range(min(10, len(subjects))))  # first 10 subjects by default

    num_subjects = len(subject_indices)
    plt.figure(figsize=(imgs_per_subject * 2, num_subjects * 2))

    idx = 1
    for k in subject_indices:
        sid = subjects[k]
        imgs = X_blocks[k]
        nums = ids_blocks[k]
        nshow = min(imgs_per_subject, imgs.shape[0])
        for j in range(nshow):
            plt.subplot(num_subjects, imgs_per_subject, idx)
            plt.imshow(imgs[j], cmap="gray")
            plt.axis("off")
            plt.title(f"S{sid}_{nums[j]}", fontsize=9)
            idx += 1

    plt.suptitle("CMU PIE (Train)  Titles as S<sid>_<num>", y=1.02)
    plt.tight_layout()
    if save_path: plt.savefig(save_path, dpi=200)
    plt.show()

# ---------------- Main ----------------
if __name__ == "__main__":
    # 1) Pick subjects by seed (assignment rule)
    subjects = select_subjects(LAST3, k=25)
    subjects_sorted = sorted(subjects)

    # 2) Load CMU PIE blocks and split 70/30
    (Xtr_list, ytr_list, Xte_list, yte_list,
     paths_tr_list, ids_tr_list, paths_te_list, ids_te_list) = load_cmu_pie_block(
        CMU_PIE_ROOT, subjects_sorted, seed=LAST3, size=IMAGE_SIZE, train_ratio=TRAIN_RATIO
    )

    # 3) Load 10 selfies, split 7/3
    #Xself, selfie_paths = load_selfies_block(SELFIE_DIR, size=IMAGE_SIZE)
    #Xself_tr, Xself_te  = split_selfies(Xself)

    # 4) Build label map
    label_map = make_label_map(subjects_sorted)

    # 5) Relabel CMU PIE blocks & stack
    Xtr_blocks, ytr_blocks, Xte_blocks, yte_blocks = [], [], [], []
    for sid, Xtr_s, ytr_s, Xte_s, yte_s in zip(subjects_sorted, Xtr_list, ytr_list, Xte_list, yte_list):
        lab = label_map[f"subject_{sid:02d}"]
        ytr_blocks.append(relabel(ytr_s, lab))
        yte_blocks.append(relabel(yte_s, lab))
        Xtr_blocks.append(Xtr_s)
        Xte_blocks.append(Xte_s)

    # Add selfie as class 25
   # yself_tr = np.full((Xself_tr.shape[0],), label_map["selfie"], dtype=int)
   # yself_te = np.full((Xself_te.shape[0],), label_map["selfie"], dtype=int)
   # Xtr_blocks.append(Xself_tr); ytr_blocks.append(yself_tr)
   # Xte_blocks.append(Xself_te); yte_blocks.append(yself_te)

    # 6) Final tensors
    X_train, y_train = stack_xy(Xtr_blocks, ytr_blocks)   # [Ntr,32,32], [Ntr]
    X_test,  y_test  = stack_xy(Xte_blocks, yte_blocks)   # [Nte,32,32], [Nte]

    # 7) Print 26 categories
    print("== 26 Categories ==")
    for k, v in label_map.items():
        print(f"{v:2d} : {k}")
    print(f"\nTrain size: {X_train.shape[0]}   Test size: {X_test.shape[0]}")

    # 8) Plot & save selfie grid
    #show_selfies_grid(Xself, selfie_paths, cols=5, save_path=SAVE_DIR / "selfies_grid.png")

    # 9) Plot & save CMU PIE examples (train) titled S<sid>_<num>
    #    Default: first 10 subjects, 5 images each
    show_cmu_pie_examples_labeled(
        Xtr_list, subjects_sorted, ids_tr_list,
        subject_indices=list(range(min(10, len(subjects_sorted)))),
        imgs_per_subject=5,
        save_path=SAVE_DIR / "cmu_pie_examples.png"
    )

    # 10) (Optional) save arrays for later stages
    # np.savez(SAVE_DIR / "cmu_pie_split.npz",
    #          X_train=X_train, y_train=y_train, X_test=X_test, y_test=y_test, label_map=label_map)

    print(f"\nSaved figures to: {SAVE_DIR.resolve()}")
```
